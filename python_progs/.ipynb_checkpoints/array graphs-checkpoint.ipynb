```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#Draw a line in a diagram from position (0,0)to position(6,250)
xpoints = np.array([0,6])
ypoints = np.array([0,250])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
#Add title, lables, markers and grid lines to the plot : 
x = np.array([80,85,90,95,100,105,110,115,120,125])
y = np.array([240,250,260,270,280,290,300,310,320,330])
plt.title("Sport's watch data")
plt.xlabel("Average pulse")
plt.ylabel("Calorie burnage")
plt.plot(x,y,marker='o')
plt.grid()
plt.show()
```

```python
#Draw a line in a diagram from position (1,3) to (2,8) then to (6,1) and finally to(8,10)
xpoints = np.array([1,2,6,8])
ypoints = np.array([3,8,1,10])
plt.plot(xpoints,ypoints)
plt.show()
```

```python
#Draw 4 bars :
x = np.array(['A','B','C','D'])    #qualitative factor
y = np.array([3,8,1,10])         #quantitative factor
plt.bar(x,y)
plt.show()
```

```python
#Draw 4 horizontal bars :
qual = np.array(['A','B','C','D'])    #qualitative factor
quan = np.array([3,8,1,10])         #quantitative factor
plt.barh(x,y)
plt.show()
```

```python
#A simple pie chart :
data = np.array([10,20,20,15])
plt.pie(data)
plt.show()
```

```python
#A simple histogram :
x = np.random.normal(170,10,250)    #normal distribution
#print(x)
plt.hist(x)
plt.show()
```
