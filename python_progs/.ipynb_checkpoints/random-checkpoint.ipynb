```python
from random import seed
from random import random
seed(1)      #changing seed no. takes different random nos.
print(random(),random(),random())
```

```python
from random import randint
seed(2)
for i in range(7):
    value= randint(0,10)
    print(value)
```

```python
from random import gauss  # follows gaussian probablity
seed(1)
for i in range(10):
    value=gauss(0,10)
    print(value)
```

```python
seed(1)
list1=[i for i in range(10)]          # i gets inserted in the list
print(list1)
```

```python
from random import sample
seed(1)
list1=[i for i in range(10)]  
print(list1)
seed(1)
sample1=sample(list1,3)   #takes 3 sample from the list
print(sample1)
seed(2)                           #changing seed takes different random samples
sample2=sample(list1,3)
print(sample2)
```

```python
from numpy.random import seed
from numpy.random import randn
from matplotlib import pyplot
seed(1)
data1=randn(100)     #creates n d array
print(data1)
pyplot.hist(data1)
pyplot.show()
```

```python
from numpy import mean
from numpy import median
from numpy import var      #variance
from numpy import std      #standard deviation
seed(1)
data1=randn(10000)
print(data1)
pyplot.hist(data1)
pyplot.show()
result1=mean(data1)
print(result1)
result2=median(data1)
print(result2)
result3=var(data1)
print(result3)
result4=std(data1)
print(result4)
```
