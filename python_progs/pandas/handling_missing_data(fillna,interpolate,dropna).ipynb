# Handling missing data - fillna, interpolate, dropna

```python
import pandas as pd
import os
os.chdir("/home/shrishjay/programs/datasets/")
```

```python
df = pd.read_csv("weather_data_new.csv", parse_dates = ["day"])      # as the days are recognized as strings so we changed the string type to date format
print(type(df.day[0]))
df
```

```python
df.set_index(['day'], inplace = True)
df
```

## fillna

### fill all NaN with one specific value

```python
new_df = df.fillna(0)
new_df
```

### fillna with column names and dict

```python
new_df = df.fillna({'temperature' : 0, 'windspeed' : 0, 'event' : 'No event'})
new_df
```

### Use method to determine how to fill na values
### forward fill to fill the na values from the previous value

```python
df_new = df.fillna(method = 'ffill')      #this is more relevant than putting 0
df_new
```

### bfill to fill the na values from the next value

```python
df_new = df.fillna(method = 'bfill')
df_new
```

### Use of axis

```python
df_new = df.fillna(method = 'bfill', axis = 'columns')
df_new
```

### Use of limit parameter

```python
df_new = df.fillna(method = 'ffill', limit = 1)
df_new
```

### interpolate

```python
df_new = df.interpolate()
df_new
```
